```python
import requests
from bs4 import BeautifulSoup

import random
import re
import pandas as pd
import numpy as np

from datetime import datetime
```

```python
wiki_link = 'https://kworb.net/spotify/listeners.html'
#i think monthly listeners is a better metric to see whos trending now

headers = {
            'User-Agent': random.choice([
                'Mozilla/5.0 (Windows NT 10.0; Win64; x64)',
                'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_7)',
                'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:89.0)',
            ])
        }

html_text = requests.get(wiki_link, headers=headers).text
soup = BeautifulSoup(html_text, 'lxml')
```

```python
soup.select('tbody')[0].select('tr')[0].select('td')[0].text
#artist name
```

```python
soup.select('tbody')[0].select('tr')[0].select('td')[1].text
#monthly listeners
```

```python
table = soup.select('tbody')[0].select('tr')
```

```python

data = {}
for rank, row in enumerate(table[:500]):
    artist = row.select('td')[0].text.encode("latin-1").decode("utf-8")
    data[artist] = {}
    data[artist]["ranking"] = int(rank + 1)
    data[artist]["monthly_listners"] = int(str.replace(row.select('td')[1].text, ",", ""))

#can use spotify api to get spotify id
```
